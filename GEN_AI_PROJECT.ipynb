```python
import pandas as pd

# Load the dataset
file_path = "medical reports.xlsx"
df = pd.read_excel('/content/medical reports.xlsx')

# Show the columns
print("Columns in the dataset:", df.columns.tolist())

# Check if there's a 'Patient Name' or 'ID' column
if 'Patient Name' in df.columns:
    group_col = 'Patient Name'
elif 'Patient ID' in df.columns:
    group_col = 'Patient ID'
else:
    group_col = None

# Summarizer function
def generate_summary(group):
    # Use group.name if available, otherwise provide a default
    summary = f"Summary for {group.name if hasattr(group, 'name') else 'Overall'}:\n"
    for col in group.columns:
        if pd.api.types.is_numeric_dtype(group[col]):
            avg = group[col].mean()
            summary += f"  - Average {col}: {avg:.2f}\n"
        elif group[col].nunique() == 1:
            summary += f"  - {col}: {group[col].iloc[0]}\n"
    return summary

# Apply summarizer
if group_col:
    grouped = df.groupby(group_col)
    for name, group in grouped:
        print(generate_summary(group))
        print("-" * 40)
else:
    print("No patient identifier column found. Showing overall summary:\n")
    print(generate_summary(df))
```
